## United States-Scope Data

Additional FIPS specifiers are not required if a user wishes to view data for collections of geographies from the `us` upper-level/parent scope.

This default behavior for querying such data raises a `UserWarning`.

```python
import acspsuedo.data as apd
from acspsuedo.api import ACS1
```

```python
df = await apd.download_by_geo_collection(
    api  = ACS1,
    year = 2023,
    dataset = 'B25058',
    geography = 'state',
)

df.head(5)
```

Warnings can be toggled off with `ignore_warning`.

```python
df = await apd.download_by_geo_collection(
    api  = ACS1,
    year = 2023,
    dataset = 'B25058',
    geography = 'state',
    ignore_warning = True
)

df.head(5)
```
